#convert data to csv  (convert BDNews to csv first)

```python
from bs4 import BeautifulSoup
import glob
import csv
import pandas as pd

header = ['DOCNO', 'TITLE', 'TEXT']

with open('/content/data.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
  

files = []
for subfolder in glob.glob('/content/FIRE/adhoc.fire.en/en.docs.2011/en_BDNews24/*'):
  temp = subfolder + '/*'
  files.append(glob.glob(temp))


for i in range(len(files)):
  for eachpath in files[i]:
    news_file = open(eachpath)
    anchor = BeautifulSoup(news_file, "html.parser")
    datarow = [anchor.docno.text, anchor.title.text, anchor.text.strip('\n').split('\n')[3]]
    with open('/content/data.csv', 'a', encoding='UTF8') as f:
      writer = csv.writer(f)
      writer.writerow(datarow)
```

# Convert Telegraph To CSV and append to csv of BDNews

```python
from bs4 import BeautifulSoup
import glob
import csv
import pandas as pd
import os


tel_files = []
for subfolder in glob.glob('/content/FIRE/adhoc.fire.en/en.docs.2011/en_TheTelegraph_2001-2010/*'):
  temp = subfolder + '/*/*'
  tel_files.append(glob.glob(temp))

for i in range(len(tel_files)):
  for eachpath in tel_files[i]:
    if os.path.isfile(eachpath):
        news_file = open(eachpath)
        anchor = BeautifulSoup(news_file, "html.parser")
        datarow = [anchor.docno.text, 'None', anchor.text.strip('\n')]

        with open('/content/data.csv', 'a', encoding='UTF8') as f:
          writer = csv.writer(f)
          writer.writerow(datarow)
```
